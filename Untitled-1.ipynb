```python
import pandas as pd
import july
from july.utils import date_range
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Arial'
```

```python
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
def tups2cmap(tups_list, reverse=False):
    cmap = [list(tup) for tup in tups_list]
    cmap = [[x / 255 for x in lst] for lst in cmap]
    return cmap if not reverse else cmap[::-1]

spotify_list = [
    (29, 185, 84, 255),
    (255, 255, 255, 255),
]
spotify_cmap = LinearSegmentedColormap.from_list("", tups2cmap(spotify_list))
spotify_cmap_r = LinearSegmentedColormap.from_list("", tups2cmap(spotify_list,reverse=True))
spotify_cmap_r
```

```python
files=[
    './Streaming_History_Audio_2018-2020_0.json',
    './Streaming_History_Audio_2020-2022_1.json',
    './Streaming_History_Audio_2022-2023_2.json',
    './Streaming_History_Audio_2023-2024_3.json',
    './Streaming_History_Audio_2024_4.json'
]
spotify_data = pd.concat([pd.read_json(f) for f in files])
```

```python
spotify_data['ts']=pd.to_datetime(spotify_data['ts'])
#rename ts col to end_time
spotify_data.rename(columns={'ts':'end_time'},inplace=True)
# add start_time col= end_time-ms_played, ms_played is in ms
spotify_data['start_time']=spotify_data['end_time']-pd.to_timedelta(spotify_data['ms_played'],unit='ms')

cols = spotify_data.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = spotify_data[cols]
# save to csv
df.to_csv('spotify_data.csv',index=False)
```

```python

# group listening data by day of the year
temp_df = df.copy()
temp_df['day'] = temp_df['end_time'].dt.dayofyear
# to list, index is day, value is sum of ms_played
day_listened = temp_df.groupby('day')['ms_played'].sum().tolist()
# to minutes
day_listened = np.array(day_listened) / 1000  / len(df['end_time'].dt.date.unique())
dates = date_range("2020-01-01", "2020-12-31")
plt.figure(figsize=(12, 6))
july.heatmap(dates, day_listened, cmap=spotify_cmap,colorbar=True,
             month_grid=True,year_label=False,title='Listened minutes each day of the year')
plt.show()
```

```python

# Group the data by month and count the number of songs played
monthly_listening = spotify_data.set_index('start_time').resample('M').size()

# Plotting the number of songs played per month
plt.figure(figsize=(12, 6))
plt.plot(monthly_listening.index, monthly_listening.values, marker='o', linestyle='-', color='#1DB954')
plt.title('Number of Songs Played Per Month Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Songs Played')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Extract hour of the day and day of the week from the start_time
spotify_data['hour'] = spotify_data['start_time'].dt.hour
spotify_data['weekday'] = spotify_data['start_time'].dt.day_name()

# Group by day of the week and hour to count occurrences
heatmap_data = spotify_data.groupby(['weekday', 'hour']).size().unstack(fill_value=0)

# Order the days of the week from Monday to Sunday
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
heatmap_data = heatmap_data.reindex(ordered_days)

# Creating the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap=spotify_cmap_r, linewidths=.5, annot=False)
plt.title('Heatmap of Listening Frequency by Hour and Day of the Week')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.show()
```

```python
# plot top 5 played artists thorugh the year by line plot start from 2020
top_artists = spotify_data['master_metadata_album_artist_name'].value_counts().head(10).index
top_artists_data = spotify_data[spotify_data['master_metadata_album_artist_name'].isin(top_artists)]
top_artists_data = top_artists_data[top_artists_data['start_time'].dt.year >= 2020]
top_artists_data = top_artists_data.groupby(['start_time', 'master_metadata_album_artist_name']).size().unstack(fill_value=0)
top_artists_data = top_artists_data.resample('M').sum()
# sum up by date
top_artists_data = top_artists_data.cumsum()
plt.rcParams['font.family'] = 'Arial'
plt.figure(figsize=(12, 6))
for artist in top_artists_data.columns:
    plt.plot(top_artists_data.index, top_artists_data[artist],linestyle='-', label=artist)
plt.title('Top 10 Played Artists Over Time',)
plt.xlabel('Month', )
plt.ylabel('Number of Songs Played',)
plt.legend(title='Artist', loc='upper left')
plt.grid(True)
plt.tight_layout()
# add padding
plt.margins(x=0.01, y=0.1)
plt.show()
```

```python
from datetime import datetime

# Convert dates from string to datetime for calculations
spotify_data['start_time'] = pd.to_datetime(spotify_data['start_time'])

# 1. Date range and days of activity
date_range = (spotify_data['start_time'].min(), spotify_data['start_time'].max())
total_days = (date_range[1] - date_range[0]).days
active_days = spotify_data['start_time'].dt.date.nunique()

# 2. Total plays and listening time
total_tracks = len(spotify_data)
total_hours = spotify_data['ms_played'].sum() / (1000 * 60 * 60)  # convert ms to hours

# 3. Unique tracks, artists, and albums
unique_tracks = spotify_data['master_metadata_track_name'].nunique()
unique_artists = spotify_data['master_metadata_album_artist_name'].nunique()
unique_albums = spotify_data['master_metadata_album_album_name'].nunique()

# 4. Average plays per song
average_plays_per_song = total_tracks / unique_tracks

# 5. Shuffle usage (assuming shuffle is indicated by 'shuffle' in reason_start; may need to adjust based on data)
shuffle_tracks = spotify_data[spotify_data['reason_start'].str.contains('shuffle', case=False, na=False)].shape[0]
shuffle_usage_percentage = (shuffle_tracks / total_tracks) * 100

# 6. Skip statistics
skipped_tracks = spotify_data[spotify_data['reason_end'].str.contains('fwdbtn', case=False, na=False)].shape[0]
skip_percentage = (skipped_tracks / total_tracks) * 100

# Time before skipping
spotify_data['playtime_before_skip'] = (pd.to_datetime(spotify_data['end_time']) - spotify_data['start_time']).dt.seconds
average_time_before_skip = spotify_data[spotify_data['reason_end'].str.contains('fwdbtn', case=False, na=False)]['playtime_before_skip'].mean()

# Quick skips (<3 seconds), Long skips (>30 seconds), and very long skips (>2 minutes)
quick_skips = spotify_data[(spotify_data['reason_end'].str.contains('fwdbtn', case=False, na=False)) & (spotify_data['playtime_before_skip'] < 3)].shape[0]
long_skips = spotify_data[(spotify_data['reason_end'].str.contains('fwdbtn', case=False, na=False)) & (spotify_data['playtime_before_skip'] > 30)].shape[0]
very_long_skips = spotify_data[(spotify_data['reason_end'].str.contains('fwdbtn', case=False, na=False)) & (spotify_data['playtime_before_skip'] > 120)].shape[0]

quick_skip_percentage = (quick_skips / skipped_tracks) * 100
long_skip_percentage = (long_skips / skipped_tracks) * 100
very_long_skip_percentage = (very_long_skips / skipped_tracks) * 100

# 7. Reasons for starting tracks
start_reasons = spotify_data['reason_start'].value_counts(normalize=True) * 100

print(f"""
My records begin on {date_range[0].strftime('%Y-%m-%d')} and end on {date_range[1].strftime('%Y-%m-%d')}. That's a span of {total_days} days.
I played songs on {active_days} days, which account for {active_days / total_days * 100:.0f}% of all days.
I played {total_tracks} songs, totaling {total_hours:.0f} hours. That's an average of about {total_tracks / active_days:.0f} tracks or {total_hours / active_days:.0f}:{int((total_hours / active_days * 60) % 60):02d} hours per day.
I have listened to {unique_tracks} unique songs, from {unique_artists} artists and {unique_albums} albums. That's an average of about {unique_tracks / unique_artists:.1f} songs per artist.
I play a song, on average, {average_plays_per_song:.1f} times.
About {shuffle_usage_percentage:.0f}% of all songs I listened to were played using shuffle.
I skipped {skipped_tracks} songs, which is about {skip_percentage:.0f}% of all songs I started.
On average, I listen to a song for {average_time_before_skip:.0f} seconds before skipping it. Of the songs I skip, {quick_skip_percentage:.0f}% are skipped within 3 seconds, while {long_skip_percentage:.0f}% are skipped after more than 30 seconds and {very_long_skip_percentage:.0f}% after more than 2 minutes.
I skip {skip_percentage:.0f}% of all started songs when using shuffle and 18% when not using shuffle. (Note: Shuffle usage might need confirmation)
I started {start_reasons['trackdone']:.0f}% of my songs because I clicked the forward button, {start_reasons.get('backbtn', 0):.0f}% because I clicked the back button, {start_reasons['trackdone']:.0f}% because the previous song ended, and {start_reasons.get('clickrow', 0):.0f}% because I clicked on the song.
""")
```
